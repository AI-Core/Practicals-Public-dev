## Example with heights

```python
avg_height = 170
std_height = 10
np.random.seed(1)
data = np.random.normal(loc=avg_height, scale=std_height, size=20000)
count, bins, patches = plt.hist(data, 30, density=True)
plt.plot(bins, 1/(std_height * np.sqrt(2 * np.pi)) *

               np.exp( - (bins - avg_height)**2 / (2 * std_height **2) ),

         linewidth=2, color='r')
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
import scipy.stats as st
from scipy.stats import zmap
z_180 = (180 - avg_height) / std_height
p_180=st.norm.cdf(z_180)
z_160 = (160 - avg_height) / std_height
p_160=st.norm.cdf(z_160)
p = p_180 - p_160
print(f'The probability of a person being below 180 cm is: {p_180}')
print(f'The probability of a person being below 160 cm is: {p_160}')
print(f'The probability of a person being between 160 cm and 180 cm is: {p}')

count, bins, patches = plt.hist(data, 100, density=True)
lower_bins = [i for i, x in enumerate(bins <= 160) if x]
upper_bins = [i for i, x in enumerate(bins > 180) if x]
for i in lower_bins:
    patches[i].set_facecolor('w')
for i in upper_bins:
    patches[i - 1].set_facecolor('w')
plt.plot(bins, 1/(std_height * np.sqrt(2 * np.pi)) *

               np.exp(-(bins - avg_height)**2 / (2 * std_height **2) ),

         linewidth=2, color='r')
plt.xlabel('Height (cm)')
plt.ylabel('Probability')
plt.title('Height Distribution')
plt.show()
```

## Significance level  

```python
import scipy.stats as st
z_upper = st.norm.ppf(0.975) # z-value corresponding to 97.5% confidence
z_lower = st.norm.ppf(0.025) # z-value corresponding to 2.5% confidence
upper = z_upper * std_height + avg_height
lower = z_lower * std_height + avg_height
print('The upper bound is:', upper)
print('The lower bound is:', lower)
```

We have a 5% chances of being above `upper` or below `lower`

```python
count, bins, patches = plt.hist(data, 30, density=True)
lower_bins = [i for i, x in enumerate(bins < lower) if x]
upper_bins = [i for i, x in enumerate(bins > upper) if x]
for i in lower_bins:
    patches[i].set_facecolor('r')
for i in upper_bins:
    patches[i - 1].set_facecolor('r')
plt.plot(bins, 1/(std_height * np.sqrt(2 * np.pi)) *

               np.exp(-(bins - avg_height)**2 / (2 * std_height **2) ),

         linewidth=2, color='r')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Height Distribution')
plt.show()
```

## t-test

```python
data_170 = np.random.normal(loc=170, scale=10, size=100)
data_190 = np.random.normal(loc=190, scale=5, size=100)

plt.hist(data_170, 10, density=True, alpha=0.5, label='170')
plt.hist(data_190, 10, density=True, alpha=0.5, label='190')
plt.show()
```

```python
import scipy.stats as st
t_stat, p_val= st.ttest_ind(data_190, data_170)
print(f'The t-test value is {t_stat}, and the p-value is {p_val}')
```

```python
import scipy.stats as st
data_170 = np.random.normal(loc=170, scale=5, size=50)
data_175 = np.random.normal(loc=173, scale=5, size=50)

plt.hist(data_170, 5, density=True, alpha=0.5, label='170')
plt.hist(data_175, 5, density=True, alpha=0.5, label='175')
plt.show()
t_stat, p_val= st.ttest_ind(data_175, data_170)
print(f'The t-test value is {t_stat}, and the p-value is {p_val}')
```

## A/B test

```python
np.random.seed(0)
data_A = np.random.normal(loc=0.15, scale=5, size=30)
data_B = np.random.normal(loc=0.20, scale=7, size=30)
plt.hist(data_A, 5, density=True, alpha=0.5, label='A')
plt.hist(data_B, 5, density=True, alpha=0.5, label='B')
plt.legend()
plt.show()
```

```python
t_stat, p_val= st.ttest_ind(data_A, data_B)
print(f'The t-test value is {t_stat}, and the p-value is {p_val}')
```
